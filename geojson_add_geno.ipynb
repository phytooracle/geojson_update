```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv('/home/emmanuelgonzalez/Gantry_fieldbook_2020_Revised_Irr_TRT.csv', dtype='str').set_index('Plot')
```

```python
gj = gpd.read_file('/home/emmanuelgonzalez/season11_multi_latlon.geojson')
```

```python
gj['genotype'] = None

for i, row in gj.iterrows():
    #row['genotype'] = None
    plot = row['ID']
    row_df = df.loc[plot]
    geno = row_df['PI_accession']
    #print(plot, geno)
    gj['genotype'][i] = str(geno)
```

```python
gj.to_file('/home/emmanuelgonzalez/season11_multi_latlon_geno.geojson', driver='GeoJSON')
```
